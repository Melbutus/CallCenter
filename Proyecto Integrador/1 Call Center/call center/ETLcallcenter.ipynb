```python
import pandas as pd
import numpy as np
import string
import random
import uuid
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# chequeo de ceros, espacios y nulos en las columnas
def verificar_valores(df, columna): # función para verificar valores nulos, ceros y espacios en blanco
    nulos = df[columna].isnull().sum()
    ceros = (df[columna] == 0).sum()
    espacios = (df[columna].astype(str).str.strip() == '').sum()
    return nulos, ceros, espacios

columnas = [ #columnas a verificar
    'call_id', 'customer_id', 'vru.line', 'priority', 'type', 'date',
    'vru_entry', 'vru_exit', 'vru_time', 'q_start', 'q_exit', 'q_time',
    'outcome', 'ser_start', 'ser_exit', 'ser_time', 'server', 'startdate'
]

for columna in columnas: # Iterar sobre las columnas y aplicar la función de verificación
    nulos, ceros, espacios = verificar_valores(archivo, columna)
    print(f'nulos en "{columna}": {nulos}, valores cero: {ceros}, en blanco o espacios: {espacios}')
```

```python
# chequeo de duplicados en las columnas 
columnas = ['call_id', 'customer_id']

for columna in columnas:
    num_duplicados = archivo.duplicated(subset=[columna]).sum()
    print(f"Número de duplicados en '{columna}':", num_duplicados)

for columna in columnas: # frecuencia de duplicados en cada columna
    conteo_duplicados = archivo[columna].value_counts()
    conteo_duplicados = conteo_duplicados[conteo_duplicados > 1]
    print(f"Conteo de duplicados en '{columna}':")
    print(conteo_duplicados)
```

```python
# Convertir `call_id` a texto y eliminar espacios #id de llamada es unico?
archivo['call_id'] = archivo['call_id'].astype(str).str.strip()

# Identificar IDs con longitud incorrecta
ids_incorrectos = archivo[archivo['call_id'].apply(lambda x: len(x) != 5)]
print("IDs con longitud incorrecta:")
print(ids_incorrectos[['call_id']])

# la columna 'call_id' está en formato string y sin espacios
archivo['call_id'] = archivo['call_id'].astype(str).str.strip()

duplicados = archivo[archivo.duplicated('call_id', keep=False)] # Verificar duplicados en 'call_id'

print(f"Cantidad de IDs duplicados: {len(duplicados)}")
print("IDs duplicados:\n", duplicados['call_id'].unique())

def generate_unique_id(existing_ids): #  generar IDs únicos de longitud 5
    while True:
        new_id = str(uuid.uuid4().hex[:5].upper())
        if new_id not in existing_ids:
            existing_ids.add(new_id)
            return new_id

existing_ids = set(archivo['call_id'])

duplicados_indices = duplicados.index # Generar nuevos IDs únicos para los duplicados
num_duplicados = len(duplicados_indices)
new_ids = [generate_unique_id(existing_ids) for _ in range(num_duplicados)]

archivo.loc[duplicados_indices, 'call_id'] = new_ids # Reemplazar los IDs duplicados con los nuevos

duplicados_ids = archivo[archivo.duplicated('call_id', keep=False)]['call_id'].unique()
if not duplicados_ids.size:
    print("No se encontraron IDs duplicados después del reemplazo.")
else:
    print('IDs duplicados:', duplicados_ids)
```

```python
# Convertir 'customer_id' a texto y reemplazar '0' y '0.0' por 'No reconocido'
archivo['customer_id'] = archivo['customer_id'].astype(str).replace(['0', '0.0'], 'No reconocido')

zero_count = (archivo['customer_id'] == 'No reconocido').sum() #cuántos registros tienen 'No reconocido'
print(f"Cantidad de registros con 'No reconocido': {zero_count}")

print(archivo['customer_id'].describe())
```

```python
# customer_id chequeamos que tengan hasta 12 digitos y que no haya duplicados
# Convertir a texto para asegurar consistencia
archivo['customer_id'] = archivo['customer_id'].astype(str).str.strip()

def generate_unique_id(existing_ids): # generar un ID único con longitud de hasta 12 dígitos
    while True:
        new_id = ''.join(random.choices(string.digits, k=12))  
        if new_id not in existing_ids:
            return new_id

valid_ids = archivo[archivo['customer_id'] != 'No reconocido']
archivo['longitud_customer_id'] = archivo['customer_id'].str.len() # Identificar IDs con longitud incorrecta
invalid_length_ids = archivo[(archivo['longitud_customer_id'] > 12) & (archivo['customer_id'] != 'No reconocido')]

existing_ids = set(valid_ids['customer_id']) # Generar nuevos IDs únicos para los registros con longitud incorrecta
new_ids = [generate_unique_id(existing_ids) for _ in range(len(invalid_length_ids))]

archivo.loc[invalid_length_ids.index, 'customer_id'] = new_ids # Asignar nuevos IDs a los registros con longitud incorrecta

# Verificar si los customer_id tienen una longitud fuera del rango 0-12
invalid_length_ids = archivo[archivo['longitud_customer_id'] > 12]  
print("IDs con longitud fuera del rango 0-12:")
print(invalid_length_ids[['customer_id', 'longitud_customer_id']])

archivo.drop(columns=['longitud_customer_id'], inplace=True)
```

```python
#  PRIORITY
# cambiar q_time 0= no esperaron?
def time_to_seconds(time_str):
    # Convertir formato HH:MM:SS a segundos, maneja '00:00:00' y valores nulos
    if pd.isna(time_str) or time_str == '00:00:00':
        return 0
    try:
        h, m, s = map(int, time_str.split(':'))
        return h * 3600 + m * 60 + s
    except ValueError:
        return 0

def seconds_to_time(seconds):  # Convierte segundos a formato HH:MM:SS
    if pd.isna(seconds) or seconds == 0:
        return "00:00:00"
    hours, remainder = divmod(int(seconds), 3600)
    minutes, seconds = divmod(remainder, 60)
    return f"{hours:02}:{minutes:02}:{seconds:02}"

archivo['q_start_seconds'] = archivo['q_start'].apply(time_to_seconds)
archivo['q_exit_seconds'] = archivo['q_exit'].apply(time_to_seconds)

# Calcular q_time en segundos
archivo['q_time'] = archivo['q_exit_seconds'] - archivo['q_start_seconds']

# q_start y q_exit en formato HH:MM:SS para la visualización
archivo['q_start'] = archivo['q_start_seconds'].apply(seconds_to_time)
archivo['q_exit'] = archivo['q_exit_seconds'].apply(seconds_to_time)

archivo.drop(columns=['q_start_seconds', 'q_exit_seconds'], inplace=True) # Eliminar las columnas intermedias 

print(archivo[['q_start', 'q_exit', 'q_time']].head(30))
```

```python
# Verificar valores válidos en `outcome` # chequear
'''valid_outcomes = ['AGENT', 'HANG', 'PHANTOM']
archivo = archivo[archivo['outcome'].isin(valid_outcomes)]
archivo['outcome'].head()

archivo[['ser_start','ser_time','ser_exit']].info()'''

# Verificar y corregir ser_start y ser_exit
def format_time(value):
    try:
        if len(value) == 8 and value.count(':') == 2:  # Formato HH:MM:SS
            return value
        # Convertir a HH:MM:SS si el valor es en segundos
        value = int(value)
        hours, remainder = divmod(value, 3600)
        minutes, seconds = divmod(remainder, 60)
        return f"{hours:02}:{minutes:02}:{seconds:02}"
    except ValueError:
        return "00:00:00"  # Valor por defecto en caso de error

archivo['ser_start'] = archivo['ser_start'].apply(format_time)
archivo['ser_exit'] = archivo['ser_exit'].apply(format_time)

invalid_ser_start = archivo[archivo['ser_start'].str.len() != 8] # Verificar la corrección
invalid_ser_exit = archivo[archivo['ser_exit'].str.len() != 8]

def time_to_seconds(time_str): # Calcular y verificar ser_time
    h, m, s = map(int, time_str.split(':'))
    return h * 3600 + m * 60 + s

archivo['ser_start_seconds'] = archivo['ser_start'].apply(time_to_seconds)
archivo['ser_exit_seconds'] = archivo['ser_exit'].apply(time_to_seconds)

archivo['ser_time'] = archivo['ser_exit_seconds'] - archivo['ser_start_seconds'] # Calcular ser_time

invalid_ser_time = archivo[(archivo['ser_time'] < 0) | (archivo['ser_time'] > 999)] # Verificar ser_time
print("entradas ser_time incorrectas:", invalid_ser_time[['ser_start', 'ser_exit', 'ser_time']])

archivo = archivo[(archivo['ser_time'] >= 0) & (archivo['ser_time'] <= 999)] # Eliminar registros anómalos

archivo.drop(columns=['ser_start_seconds', 'ser_exit_seconds'], inplace=True) # Eliminar las columnas temporales

count_invalid_ser_time = invalid_ser_time['ser_time'].count() # Verificar entradas con más de 3 dígitos en ser_time
print("Entradas con más de 3 dígitos:", count_invalid_ser_time)
```

```python
# Renombrar los tipos de servicio utilizando el método .map()
tipo_servicio = {
    'PS': 'Actividad Regular',
    'PE': 'Actividad Regular',
    'IN': 'Consulta por internet',
    'NE': 'Actividad por Acciones',
    'NW': 'Cliente potencial',
    'TT': 'Devolución de llamada fallida'
}


if 'tipo' in archivo.columns:
    archivo['tipo'] = archivo['tipo'].map(tipo_servicio) #mapea, aplica cambios

    print("Valores únicos en 'tipo' después del mapeo:\n", archivo['tipo'].unique())   # Verifica
```
